<h4 Style= 'color: #f08309' >2.2 Contar la cantidad de valores faltantes por variable (Columnas) </h4> 

```python
dfTemp = dfAutos.isnull()
dfTemp.head(5)
```

```python
# Ciclo For para contar los valores False y True (valores faltantes) de cada una de las columnas del dataFrama
for columna in dfTemp.columns.values.tolist():
    print(columna)
    print(dfTemp[columna].value_counts())
    print("")
```

<h4 Style= 'color: #f08309' >2.3 Imputación simple de valores númericos </h4> 

```python
# Para calcular el promedio de la columna y reemplazar los valores NaN con ese valor.
promNormalizadas= dfAutos["normalized"].astype("float").mean(axis=0)
promNormalizadas
```

```python
dfAutos["normalized"].replace(np.nan, 122, inplace=True)
dfAutos.head(5)
```

```python
# Promedio de variable bore y reemplazo de valores NaN
promBore= dfAutos["bore"].astype("float").mean(axis=0)
dfAutos["bore"].replace(np.nan, promBore, inplace=True)

# Promedio de variable stroke y reemplazo de valores NaN
promStroke= dfAutos["stroke"].astype("float").mean(axis=0)
dfAutos["stroke"].replace(np.nan, promStroke , inplace=True)

# Promedio de variable horsepower y reemplazo de valores NaN
promHorse= dfAutos["horsepower"].astype("float").mean(axis=0)
dfAutos["horsepower"].replace(np.nan, promHorse , inplace=True)

# Promedio de variable peak y reemplazo de valores NaN
promPeak= dfAutos["peak"].astype("float").mean(axis=0)
dfAutos["peak"].replace(np.nan, promPeak , inplace=True)

dfAutos.head(5)
```

<h4 Style= 'color: #f08309' >2.4 Imputación simple de valores categóricos </h4> 

```python
# Contar los valores según la categoria para que devuelva el valor más frecuente

dfAutos["doors"].value_counts().idxmax()
```

```python
dfAutos["doors"].replace(np.nan, "four", inplace=True)
```

<h4 Style= 'color: #f08309' >2.5 Eliminar filas con valores faltantes </h4> 

```python
dfAutos.dropna(subset = ["price"], axis =0, inplace = True)
dfAutos.reset_index(drop =True, inplace =True)
```

```python
# bucle for para contar si hay valores faltantes (NaN)

dfTemp =dfAutos.isnull()
for columna in dfTemp.columns.values.tolist():
    print(columna)
    print(dfTemp[columna].value_counts())
    print("")
```

```python
dfAutos.dtypes
```

```python
dfAutos.head(5)
```

```python
# Para convertir el tipo de dato a "int" o "float"

dfAutos[["normalized", "horsepower", "peak", "price"]] = dfAutos[["normalized", "horsepower", "peak", "price"]].astype ("int") 

dfAutos[["bore", "stroke"]] = dfAutos[["bore", "stroke"]].astype ("float") 
dfAutos.dtypes
```

<h4 Style= 'color: #f08309' >3.2 Estandarizar los datos</h4> 

```python
dfAutos.head(3)
```

```python
#Agregar una nueva columna calculada con las medidas estándar de interpretación del contexto 

dfAutos["City L/100km (calc)"] = 235/dfAutos["city"]
dfAutos["Higway L/100km (calc)"] = 235/dfAutos["higway"]

dfAutos.head(3)
```

<h4 Style= 'color: #f08309' >3.3 Volver variables comparables</h4> 

```python
dfTemp=dfAutos[["horsepower", "peak"]]
dfTemp.head(3)
```

```python
dfAutos["horsepower (norm)"] =dfAutos["horsepower"]/dfAutos ["horsepower"].max()
dfAutos["peak (norm)"] =dfAutos["horsepower"]/dfAutos ["peak"].max()
dfAutos.head(5)
```

<h4 Style= 'color: #f08309' >3.4 Binning</h4> 

```python
dfTemp.head(5)
```

```python
# Para saber el valor máximo en la columna "horsepower"

max(dfAutos["horsepower"])
```

```python
# Para saber los puntos referenciales de la columna

bins =np.linspace(48,262,4)
bins
```

```python
# Creación y aplicación de categorias para crear rangos en función de los puntos de referencia de la columna

categoria=["bajo", "medio", "alto"]
dfAutos["Horsepower(Bins)"]=pd.cut(dfAutos["horsepower"],bins,labels=categoria, include_lowest=True)

dfAutos[["horsepower", "Horsepower(Bins)"]].head(5)
```

```python
# Crear un dataFrame para agrupar la columna "make" según fabricante
gf = dfAutos.groupby (["make"]).count()
gf = dfAutos.groupby(['make']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('symboling', ascending=False)
gf
```
